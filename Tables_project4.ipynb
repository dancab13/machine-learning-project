```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store filepaths into variable
games = Path("Data/games_updated.csv")
games_details = Path("Data/games_details_updated.csv")
players = Path("Data/players_updated.csv")
ranking = Path("Data/ranking_updated.csv")
teams = Path("Data/teams_updated.csv")
```

```python
# Read in files and make dfs
games_df = pd.read_csv(games)
games_details_df = pd.read_csv(games_details)
players_df = pd.read_csv(players)
ranking_df = pd.read_csv(ranking)
teams_df = pd.read_csv(teams)
```

```python
# Merge teams and ranking and view
merge_team_ranking_df =pd.merge(teams_df, ranking_df, on='TEAM_ID')
merge_team_ranking_df
```

```python
# Merge teams and players and view
merge_team_players_df =pd.merge(teams_df, players_df, on='TEAM_ID')
merge_team_players_df
```

```python
# Merge games and details and view
merge_games_n_details_df =pd.merge(games_df, games_details_df, on='GAME_ID')
merge_games_n_details_df
```

```python
# Merge teams and games and view
merge_teams_games_df =pd.merge(teams_df, games_df, left_on='TEAM_ID', right_on= "HOME_TEAM_ID")
merge_teams_games_df
```

```python
merge_team_players_df.to_csv('Output/team_player.CSV', encoding='utf8')
```
